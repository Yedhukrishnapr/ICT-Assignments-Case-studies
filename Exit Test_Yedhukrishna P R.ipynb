# Data Preprocessing

# Missing values detection

```python
#Checking for missing values
df.isnull().sum()
```

```python
df.columns
```

```python
# Columns with missing values
columns_with_missing = ['gender', 'weight', 'ball_controlling_skills', 'jumping_skills', 
    'penalties_conversion_rate', 'coaching', 'years_of_experience',
    'mental_strength', 'shot_accuracy', 'strong_foot', 'behaviour_rating', 
    'matches_played', 'fitness_rating']
```

```python
# Creating a SimpleImputer with strategies for numerical and categorical columns
from sklearn.impute import SimpleImputer
num_imputer = SimpleImputer(strategy='mean')
cat_imputer = SimpleImputer(strategy='most_frequent')
```

```python
# Fill missing values based on data type and characteristics
for column in columns_with_missing:
    if df[column].dtype == 'float64' or df[column].dtype == 'int64':
        # For numerical columns, filling with the mean
        df[column] = num_imputer.fit_transform(df[[column]])
    else:
        # For categorical columns, filling with the mode (most frequent value)
        df[column] = cat_imputer.fit_transform(df[[column]])
```

```python
# Check the number of missing values after imputation
df.isnull().sum()
```

# Z - Score Method

```python
# Define a function to detect outliers using Z-score
def find_outliers_zscore(df, threshold=3):
    z_scores = np.abs((df - df.mean()) / df.std())
    return z_scores > threshold
```

```python
df.columns
```

```python
# Find outliers in all columns (except the 'activityID' and 'PeopleId' columns if they exist)
outliers_all_columns = df.drop(columns=['player_id','name','gender','country','height', 'weight','strong_foot','dedication_level','coaching']).apply(find_outliers_zscore)
```

```python
# Show the rows where any column has an outlier
outliers_rows = outliers_all_columns.any(axis=1)
outliers_detected = df[outliers_rows]
```

# Encoding

```python
df.columns
```

```python
#Label Encoding
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df['country'] = le.fit_transform(df['country'])
df['player_id'] = le.fit_transform(df['player_id'])
df['name'] = le.fit_transform(df['name'])
df['dedication_level'] = le.fit_transform(df['dedication_level'])
```

```python
df.head()
```

```python
# Converting the 'weight' column to strings, remove "lbs", and convert to numeric. Because we cannot encode when numerical features are given in units(with "lbs")
df['weight'] = pd.to_numeric(df['weight'].astype(str).str.replace('lbs', ''), errors='coerce')
```

```python
# Convert the 'height' column to numeric format (in inches)
def convert_height(height_str):
    feet, inches = map(int, height_str.split("'"))
    total_inches = feet * 12 + inches
    return total_inches
```

```python
df['height'] = df['height'].apply(convert_height)
```

# Min Max Scaling

```python
x = df_encoded.drop('selection',axis=1)
```

```python
y = df_encoded['selection']
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
# Create the MinMaxScaler object
scaler = MinMaxScaler()
```

```python
# Fit and transform the data to perform Min-Max scaling
x_scaled = scaler.fit_transform(x)
```

```python
x_scaled=pd.DataFrame(x_scaled,columns=x.columns)
```

```python
x_scaled.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
# Split the data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x_scaled, y, test_size=0.2, random_state=42)
```

# Modelling using different Algorithms

# Logistic Regression model

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
```

```python
# Train the Logistic Regression model
lr_model = LogisticRegression(random_state=42)
lr_model.fit(x_train, y_train)
```

```python
# Make predictions on the test set
lr_y_pred = lr_model.predict(x_test)
```

```python
# Evaluate the model
print("Logistic Regression:")
print("Accuracy:", accuracy_score(y_test, lr_y_pred))
print(classification_report(y_test, lr_y_pred))
```

# KNN

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
# create an empty list.
metric_k=[]
neighbors=np.arange(3,15)
```

```python
# Train the KNN model
for k in neighbors:
  knn_model=KNeighborsClassifier(n_neighbors=k,metric='euclidean')
  knn_model.fit(x_train,y_train)
  # Make predictions on the test set
  knn_y_pred=knn_model.predict(x_test)
  acc=accuracy_score(y_test,knn_y_pred)
  metric_k.append(acc)
```

```python
# Lets plot a graph to find highest accuracy score of given k
plt.plot(neighbors,metric_k,'o-')
plt.xlabel('k')
plt.ylabel('Accuracy')
plt.grid()
```

```python
# Lets create knn model for 3 neighbors
knn_model=KNeighborsClassifier(n_neighbors=9,metric='euclidean')
knn_model.fit(x_train,y_train)
```

```python
# Make predictions on the test set
knn_y_pred=knn_model.predict(x_test)
```

```python
# Evaluate the model
print("KNN Model:")
print("Accuracy:", accuracy_score(y_test, knn_y_pred))
print(classification_report(y_test, knn_y_pred))
```
