# Scaling the dataset using StandardScaler

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)
```

```python
data_scaled.dtype
```

# K-means clustering

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=3)
kmeans.fit(data_scaled)
```

```python
labels = kmeans.labels_
```

```python
labels
```

```python
from sklearn.metrics import silhouette_score,completeness_score
```

```python
kmeans_silhouette = silhouette_score(data_scaled, kmeans.labels_)
```

```python
kmeans_silhouette
```

```python
# Plotting K-means clusters
plt.scatter(data_scaled[:, 0], data_scaled[:, 1], c=kmeans.labels_)
plt.title("K-means Clustering")
plt.show()
```

```python
 #Finding the optimum number of clusters
```

```python
cluster_range = range(2, 11)
```

```python
ssd = []
```

```python
for n_clusters in cluster_range:
    kmeans = KMeans(n_clusters=n_clusters)
    kmeans.fit(data_scaled)
    ssd.append(kmeans.inertia_)
```

```python
plt.plot(list(cluster_range), ssd, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Sum of Squared Distances')
plt.title('Elbow Method')
plt.show()
```

# Agglomerative Clustering

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
agg_clustering = AgglomerativeClustering(n_clusters=5)
labels = agg_clustering.fit_predict(data_scaled)
```

```python
labels
```

```python
agg_silhouette = silhouette_score(data_scaled, agg_clustering.labels_)
```

```python
agg_silhouette
```

```python
plt.scatter(data_scaled[:, 0], data_scaled[:, 1], c=agg_clustering.labels_)
plt.title("Agglomerative Clustering")
plt.show()
```

```python
from sklearn.metrics import silhouette_score
import scipy.cluster.hierarchy as sch
```

```python
model = AgglomerativeClustering()
```

```python
cluster_labels = model.fit_predict(data_scaled)
```

```python
dendrogram = sch.dendrogram(sch.linkage(data_scaled, method='ward'))
plt.xlabel('Samples')
plt.ylabel('Distance')
plt.title('Dendrogram')
plt.show()
```

```python
# Calculate silhouette scores for different numbers of clusters
max_clusters = 10  # Maximum number of clusters to evaluate
silhouette_scores = []
for n_clusters in range(2, max_clusters+1):
    model = AgglomerativeClustering(n_clusters=n_clusters)
    cluster_labels = model.fit_predict(data_scaled)
    score = silhouette_score(data_scaled, cluster_labels)
    silhouette_scores.append(score)
```

```python
# Plot the silhouette scores
plt.plot(range(2, max_clusters+1), silhouette_scores, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score vs. Number of Clusters')
plt.show()
```

# dB scan clustering

```python
from sklearn.cluster import DBSCAN
```

```python
# Creat the DBSCAN object
dbscan = DBSCAN(eps=3,min_samples=25)
```

```python
dbscan.fit(data_scaled)
```

```python
labels = dbscan.labels_
```

```python
n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
```

```python
print("Number of clusters:", n_clusters)
```

```python
from sklearn.decomposition import PCA
```

```python
# Perform PCA for dimensionality reduction
pca = PCA(n_components=2)
reduced_data = pca.fit_transform(data_scaled)
```

```python
# Plot the reduced data
plt.scatter(reduced_data[:, 0], reduced_data[:, 1])
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('Scatter Plot of Reduced Data')
plt.show()
```
